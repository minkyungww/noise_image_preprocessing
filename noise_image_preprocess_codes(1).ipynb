```python
##############this code is to just check number of images in each of the folder (animal classes.)
import os

#path for validation dataset
validation_path = "/content/validation_data"

#checking if files in each folder are either jpg or png
def check_image_extensions(path, extensions=('.jpg', '.png')):
    class_folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]
    valid_extensions = set(extensions)
    results = {}

    for folder in class_folders:
        folder_path = os.path.join(path, folder)
        files = os.listdir(folder_path)
        valid_files = [file for file in files if file.lower().endswith(extensions)]
        invalid_files = [file for file in files if not file.lower().endswith(extensions)]

        ####### Add this part when desktop.ini is found
        for invalid_file in invalid_files:
            if invalid_file.lower() == 'desktop.ini':
                os.remove(os.path.join(folder_path, invalid_file))
                print(f"Deleted {invalid_file} in {folder_path}")
        #######Add this part when desktop.ini is found


        results[folder] = {
            'total_files': len(files),
            'valid_files_count': len(valid_files),
            'valid_files': valid_files,
            # 'invalid_files': invalid_files
            #######Add this part when desktop.ini is found
            'invalid_files': [file for file in invalid_files if file.lower() != 'desktop.ini']  # Exclude deleted files from results
            #######Add this part when desktop.ini is found
        }

    return results

# Run the function and print results
image_extensions_results = check_image_extensions(validation_path)
for label, data in image_extensions_results.items():
    print(f"Label: {label}")
    print(f"  Total files: {data['total_files']}")
    print(f"  Valid files count: {data['valid_files_count']}")
    if data['invalid_files']:
        print(f"  Invalid files: {data['invalid_files']}")
    else:
        print("  All files have valid extensions.")
    print("\n")
```

```python
#this is for training part
import os
import shutil
import json

# Define paths
processed_path = "/content/Processed"
imgs_path = os.path.join(processed_path, 'imgs')
json_output_path = os.path.join(processed_path, 'data-labels.json')

# Create directories if they do not exist
if not os.path.exists(processed_path):
    os.makedirs(processed_path)
if not os.path.exists(imgs_path):
    os.makedirs(imgs_path)

# Function to create label vector
def create_label_vector(label, num_classes=10):
    vector = [0] * num_classes
    vector[int(label)] = 1
    return vector

# Initialize the label map dictionary
label_map = {}

# Check if the Processed directory exists and list the contents
if os.path.exists(processed_path):
    files = os.listdir(processed_path)
    print("Sample files in 'Processed' folder:")
    for file in files[:10]:  # Print first 10 files for a quick check
        print(file)
    if len(files) > 10:
        print("... and more")

    # Process each file in the Processed folder
    for file in files:
        if file.endswith(".jpg"):
            # Extract the label from the file name (e.g., '9_img_...' -> '9')
            label = file.split('_')[0]
            img_name = file.split('_', 1)[1]
            label_map[os.path.splitext(img_name)[0]] = create_label_vector(label)

            # Move the file to the imgs folder
            source_path = os.path.join(processed_path, file)
            destination_path = os.path.join(imgs_path, img_name)
            shutil.move(source_path, destination_path)

    # Save the label map to a JSON file
    with open(json_output_path, 'w') as json_file:
        json.dump(label_map, json_file, indent=4)

    print(f"Data labels saved to {json_output_path}")

else:
    print("Directory does not exist:", processed_path)

# Verify the content of the JSON file if it exists
if os.path.exists(json_output_path):
    with open(json_output_path, 'r') as file:
        label_map = json.load(file)
    # Print part of the dictionary to understand its structure
    print("Sample entries from the 'data-labels.json' file:")
    print(dict(list(label_map.items())[:10]))  # Print first 10 entries
else:
    print(f"File does not exist: {json_output_path}")
```

```python
from PIL import Image
import os
import os.path
import torchvision.transforms as transforms
from typing import Any, Callable, Optional, Tuple
from torchvision.datasets.vision import VisionDataset
import json
import torch

class DS801Dataset(VisionDataset):

    def __init__(
            self,
            data_path: str,
            transform: Optional[Callable] = None,
    ) -> None:

        # load data
        self.data_path = data_path
        # list of file names: File name must be this string format "label-filename"
        self.file_list = os.listdir(os.path.join(data_path, 'imgs'))
        label_map_path = os.path.join(data_path, 'data-labels.json')
        self.label_map = json.load(open(label_map_path))


        self.transform = transform


    def __getitem__(self, index: int) -> Tuple[Any, Any]:
        """
        Args:
            index (int): Index

        Returns:
            tuple: (image, target) where target is index of the target class.
        """

        image_path = os.path.join(self.data_path, 'imgs', self.file_list[index])

        # load image
        img = Image.open(image_path).convert('RGB')
        # parsing label
        target = self.label_map[self.file_list[index].split('.')[0]]

        if self.transform is not None:
            img = self.transform(img)

        return index, img, torch.tensor(target)


    def __len__(self) -> int:
        return len(self.file_list)


# the function to transform image to tensors
def make_transforms():

    # we won't apply any data augmentation during the training pipeline, while you need to add augmented images into the datasets
    return transforms.Compose([
        transforms.Resize((64, 64)),
        # ToTensor applies standardization automatically when converting to Tensors.
        transforms.ToTensor(),
    ])


def build(data_path):
    return DS801Dataset(data_path=data_path, transform=make_transforms())
```

```python
import os
import shutil
import json

#define the path to the validation data
processed_path = "/content/augmented_voting"
imgs_path = os.path.join(processed_path, "imgs")

#create 'imgs' folder if it doesn't exist
if not os.path.exists(imgs_path):
    os.makedirs(imgs_path)

#categories and their order
categories = [
    "Cat", "Lynx", "Wolf", "Coyote", "Cheetah", "Jaguar",
    "Chimpanzee", "Orangutan", "Hamster", "Guinea pig"
]

#initialize the dictionary to store the mapping of filenames to one-hot labels
label_dict = {}

#process each category folder
for category in categories:
    category_path = os.path.join(processed_path, category)
    if os.path.isdir(category_path):
        category_index = categories.index(category)
        one_hot_label = [0] * len(categories)
        one_hot_label[category_index] = 1

        #process each file within the category folder
        for image_file in os.listdir(category_path):
            if image_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                #generate the dictionary key by removing the file extension
                file_name_without_ext = os.path.splitext(image_file)[0]
                label_dict[file_name_without_ext] = one_hot_label

                #move the file to the imgs folder
                source_path = os.path.join(category_path, image_file)
                destination_path = os.path.join(imgs_path, image_file)
                shutil.move(source_path, destination_path)

        #remove the empty category directory
        os.rmdir(category_path)

#write the dictionary to a JSON file in the validation data directory
json_path = os.path.join(processed_path, 'data-labels.json')
with open(json_path, 'w') as json_file:
    json.dump(label_dict, json_file, indent=4)

print(f'data-labels.json has been created at {json_path}')
print("All images moved and JSON file created successfully.")
```

```python
# Function to compute accuracy
@torch.no_grad()
def accuracy(output, target, topk=(1,)):
    """Computes the precision@k for the specified values of k"""
    if target.numel() == 0:
        return [torch.zeros([], device=output.device)]
    maxk = max(topk)
    batch_size = target.size(0)

    _, pred = output.topk(maxk, 1, True, True)
    pred = pred.t()
    correct = pred.eq(target.view(1, -1).expand_as(pred))

    res = []
    for k in topk:
        correct_k = correct[:k].view(-1).float().sum(0)
        res.append(correct_k.mul_(100.0 / batch_size))
    return res
```

```python
# Function to compute softmax with soft labels
class SoftCrossEntropyLoss(torch.nn.Module):
    def __init__(self):
        super(SoftCrossEntropyLoss, self).__init__()

    def forward(self, predictions, soft_labels):
        # predictions
        # soft_labels
        log_probs = torch.log_softmax(predictions, dim=1)
        loss = -torch.sum(soft_labels * log_probs, dim=1).mean()
        return loss
```

## Training ResNet-34 and Report the balance accuracy on validation data

We can see the the accuracy on the validation (currently, the training set) is very high (~ 96%) due to the overfitting.

```python
for epoch in range(_total_epochs):

    model.train()
    criterion.train()
    train_loss = 0.0
    train_accuracy = 0.0
    current_lr = scheduler.get_last_lr()

    for indices, images, labels in train_loader:
        images, labels = images.to(_device), labels.to(_device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        train_loss += loss.item()

        # softlabel to one-hot for computing labels
        _, one_hots = torch.max(labels, dim=1)
        _current_acc = accuracy(outputs, one_hots)[0].item()
        train_accuracy += _current_acc

    scheduler.step()
    print("[Train] LR: %.4f" % current_lr[0], "[ Epoch " + str(epoch+1) + " ] train loss:", '%.4f' % (train_loss / len(train_loader)), ",",
        "train accuracy:", '%.4f' % (train_accuracy / len(train_loader)))


    # check validation acc
    if validation_loader is not None:
        validation_accuracy = 0.0
        for indices, images, labels in validation_loader:
            images, labels = images.to(_device), labels.to(_device)
            outputs = model(images)
            # softlabel to one-hot for computing labels
            _, one_hots = torch.max(labels, dim=1)
            validation_accuracy += accuracy(outputs, one_hots)[0].item()
        validation_accuracy /= len(validation_loader)
        print("[Validation] Val ACC: %.4f" % validation_accuracy)
```

## Check the balanced accuracy of the model.

```python
def print_balanced_accuracy(net, dataloader):
  correct = 0
  total = 0

  cls_wise_correct = {}
  cls_wise_total = {}

  # Test Acc with the model at the last epoch
  with torch.no_grad():
      for _, images, labels in dataloader:
          images, labels = images.to(_device), labels.to(_device)
          outputs = net(images)
          _, predicted = torch.max(outputs.data, 1)

          # softlabel to one-hot for computing labels
          _, one_hots = torch.max(labels, dim=1)
          total += one_hots.size(0)
          correct += (predicted == one_hots).sum().item()

          # breakdown
          for _label, _correct in zip(one_hots, (predicted == one_hots)):

            if _label.item() not in cls_wise_correct:
              cls_wise_correct[_label.item()] = 0
              cls_wise_total[_label.item()] = 0
            cls_wise_total[_label.item()] += 1
            if _correct.item():
              cls_wise_correct[_label.item()] += 1

  print('Overall test accuracy": %.3f %%' % (100 * correct / float(total)))

  cls_wise_test_accuracy = {cls_id: cls_wise_correct[cls_id]/cls_wise_total[cls_id] for cls_id in cls_wise_correct.keys()}

  cls_wise_test_accuracy = dict(sorted(cls_wise_test_accuracy.items(), key=lambda item: item[1]))
  for cls_id, acc in cls_wise_test_accuracy.items():
    print('class', cls_id, ':', acc)

  # Compute balanced Accuracy on test data
  import numpy as np
  bACC = np.mean(list(cls_wise_test_accuracy.values()))
  print("Final Balanced Accuracy:", bACC)
```

```python
# Note that we are currently using 'training set' as 'validation set'
# So, all the training data were fitted during trainig, so the accuracies are very high regardless of data quality challenges
# This is why you must build your own clean validation data set!
print_balanced_accuracy(model, validation_loader)
```

```python
|from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import torch

# Class names dictionary
class_names = {
    0: "Cat",
    1: "Lynx",
    2: "Wolf",
    3: "Coyote",
    4: "Cheetah",
    5: "Jaguar",
    6: "Chimpanzee",
    7: "Orangutan",
    8: "Hamster",
    9: "Guinea pig"
}

# Function to show images
def show_images(image_indices, titles, dataset):
    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(image_indices):
        img, _ = dataset[idx][1], dataset[idx][2]
        img = img.permute(1, 2, 0).numpy()  # Convert tensor to numpy array for plotting
        plt.subplot(2, 5, i + 1)
        plt.imshow(img)
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

# Function to inspect classifications
def inspect_classifications(model, dataset, dataloader, class_to_check, num_correct, num_incorrect, device):
    model.eval()
    predictions = []
    true_labels = []
    indices_list = []

    with torch.no_grad():
        for indices, images, labels in dataloader:
            images = images.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())
            indices_list.extend(indices.cpu().numpy())

    # Convert true labels to the correct format (assuming labels are one-hot encoded)
    true_labels = [np.argmax(label) for label in true_labels]

    correct_indices = []
    incorrect_indices = []

    # Get indices of correct and incorrect classifications for the specified class
    for i in range(len(predictions)):
        if true_labels[i] == class_to_check:
            if predictions[i] == true_labels[i]:
                correct_indices.append(indices_list[i])
            else:
                incorrect_indices.append(indices_list[i])

    print(f"You chose class {class_to_check} ({class_names[class_to_check]}).")
    print(f"Checking {num_correct} correctly classified and {num_incorrect} incorrectly classified images.")
    print("Correct indices:", correct_indices)
    print("Incorrect indices:", incorrect_indices)

    # Show correctly classified images
    if len(correct_indices) > 0:
        correct_titles = [f"Correct: {class_names[class_to_check]}" for _ in range(min(num_correct, len(correct_indices)))]
        show_images(correct_indices[:num_correct], correct_titles, dataset)
    else:
        print("No correctly classified images to show.")

    # Show incorrectly classified images
    if len(incorrect_indices) > 0:
        incorrect_titles = [f"Pred: {class_names[predictions[incorrect_indices[i]]]}, True: {class_names[class_to_check]}" for i in range(min(num_incorrect, len(incorrect_indices)))]
        show_images(incorrect_indices[:num_incorrect], incorrect_titles, dataset)
    else:
        print("No incorrectly classified images to show.")

# Example usage
class_to_check = int(input("Which class do you want to check (enter number): "))
num_correct = int(input("How many correctly classified images do you want to check: "))
num_incorrect = int(input("How many incorrectly classified images do you want to check: "))

inspect_classifications(model, validation_data, validation_loader, class_to_check, num_correct, num_incorrect, _device)
```
